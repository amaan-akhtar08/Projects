```python
import pandas as pd

data = pd.read_csv("../data/marketing_data.csv")

data.head()
```

```python
data.info()
```

```python
data.rename(columns={' Income ': 'Income'}, inplace=True)
data.Income = data.Income.str.replace("$","")
data.Income= data.Income.str.replace(",","")
data.Income = data.Income.astype(float)
data.head()
```

```python
data.Dt_Customer = pd.to_datetime(data.Dt_Customer)
data.head()
```

```python
data.info()
```

```python
value_counts = data['Marital_Status'].value_counts()
print(value_counts)
```

```python
null_income_rows = data[data['Income'].isnull()]
print(null_income_rows)
```

```python
group_means = data.groupby(['Marital_Status', 'Education'])['Income'].mean().reset_index()
group_means.rename(columns={'Income': 'mean_income'}, inplace=True)
df = pd.merge(data, group_means, on=['Marital_Status', 'Education'], how='left')
df['Income'].fillna(df['mean_income'], inplace=True)
df.drop(columns=['mean_income'], inplace=True)
df.to_csv("../data/updated_file.csv", index=False)
df.head()
```

```python
null_income_rows = df[df['Income'].isnull()]
print(null_income_rows)
```

```python

df["Total_Child"] = df.Kidhome + df.Teenhome
df["Total_Num_Purchases"] = df.NumWebPurchases+df.NumStorePurchases+df.NumCatalogPurchases
df["Age"] = 2024 - df.Year_Birth
df["Total_Spend"] = df.MntWines+df.MntFruits+df.MntMeatProducts+df.MntFishProducts+df.MntSweetProducts+df.MntGoldProds
df.to_csv("../data/updated_file.csv", index=False)
df.head()
```

```python
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower_bound, upper_bound)

df_treated = df.copy()

for column in x_vars:
    df_treated[column] = cap_outliers(df_treated[column])

plt.figure(figsize=(15, 8))

count = 1
for i in x_vars:
    plt.subplot(2,2, count)
    sns.boxplot(x=df_treated[i])
    count+=1
```

## Null Hypothesis and alternate hypothesis
##### Null Hypothesis - No effect / No difference
##### Alternate Hypothesis - There is effect / difference

```python
from  scipy import stats as st
# Old people(Null Hypothesis) --> no.of store purchases <= no. of online purchases
# Old people(Alternate Hypo) --> no. of store purchases > no. of online purchases
# Assuming old people age threshold is 60

OldPerson_Store = df_treated.loc[df_treated.Age>=60,'NumStorePurchases']
OldPerson_Web = df_treated.loc[df_treated.Age>=60,'NumWebPurchases']
t_result  =  st.ttest_ind(OldPerson_Store,OldPerson_Web, alternative='greater')

statistics = t_result.statistic
p_value = t_result.pvalue

print(f"t-Statistics = {statistics}")
print(f"p-value = {p_value}")

if(p_value>0.05):
    print("\nObserved data is likely to be under Null Hypothesis")
else:
    print("\nObserved data is likely to be under Alternate Hypothesis")
```

```python
from  scipy import stats as st
# Parents(Null Hypothesis) --> no.of store purchases >= no. of online purchases
# Parents(Alternate Hypo) --> no. of store purchases < no. of online purchases

parent_store = df_treated.loc[df_treated.Total_Child>=1,'NumStorePurchases']
parent_web = df_treated.loc[df_treated.Total_Child>=1,'NumWebPurchases']
t_result  =  st.ttest_ind(parent_store,parent_web, alternative='less')

statistics = t_result.statistic
p_value = t_result.pvalue

print(f"t-Statistics = {statistics}")
print(f"p-value = {p_value}")

if(p_value>0.05):
    print("\nObserved data is likely to be under Null Hypothesis")
else:
    print("\nObserved data is likely to be under Alternate Hypothesis")
```

```python
from  scipy import stats as st
#  (Null Hypothesis) --> no.of store purchases >= other channels
# (Alternate Hypo) --> no. of store purchases < other channels

store_purchases = df_treated.loc[:,'NumStorePurchases']
web_purchases = df_treated.loc[:,'NumWebPurchases']
catalog_purchases = df_treated.loc[:,'NumCatalogPurchases']

# print(catalog_purchases)

other_purchases = web_purchases+catalog_purchases

t_result  =  st.ttest_ind(other_purchases,store_purchases, alternative='greater')

statistics = t_result.statistic
p_value = t_result.pvalue

print(f"t-Statistics = {statistics}")
print(f"p-value = {p_value}")

if(p_value>0.05):
    print("\nObserved data is likely to be under Null Hypothesis")
else:
    print("\nObserved data is likely to be under Alternate Hypothesis")
```

```python
from  scipy import stats as st

# (Null Hypothesis) --> US total purchases <= other country purchases
# (Alternate Hypo) --> US total purchases > other country purchases

purchases_world = df_treated.loc[df_treated.Country=='US','Total_Spend']
purchases_US = df_treated.loc[df_treated.Country!='US','Total_Spend']

t_result  =  st.ttest_ind(purchases_world,purchases_US, alternative='greater')

statistics = t_result.statistic
p_value = t_result.pvalue

print(f"t-Statistics = {statistics}")
print(f"p-value = {p_value}")

if(p_value>0.05):
    print("\nObserved data is likely to be under Null Hypothesis")
else:
    print("\nObserved data is likely to be under Alternate Hypothesis")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_treated.dropna(subset=['Age', 'Response'], inplace=True)
age_bins = range(df_treated['Age'].min() - 1, df_treated['Age'].max() + 11, 10)
acceptance_rate = df_treated.groupby(pd.cut(df_treated['Age'], bins=age_bins))['Response'].mean()
acceptance_rate_df = pd.DataFrame(acceptance_rate)
acceptance_rate_df.reset_index(inplace=True)
acceptance_rate_df.columns = ['Age_Range', 'Acceptance_Rate']
acceptance_rate_df['Acceptance_Rate_perc'] = acceptance_rate_df['Acceptance_Rate']*100

plt.figure(figsize=(10, 6))
sns.barplot(data=acceptance_rate_df, x='Age_Range', y='Acceptance_Rate_perc')
plt.xlabel('Age Range')
plt.ylabel('Acceptance Rate')
plt.title('Acceptance Rate by Age Range')
plt.xticks(rotation=45)
plt.ylim(0, 30)
plt.grid(True)

for index, row in acceptance_rate_df.iterrows():
    plt.text(index, float(row['Acceptance_Rate_perc']) + 0.02, f"{float(row['Acceptance_Rate_perc']):.2f}", ha='center')

plt.tight_layout()
plt.show()
```
